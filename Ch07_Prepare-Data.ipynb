```python
# load required libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import Binarizer, MinMaxScaler, Normalizer, StandardScaler
```

```python
filename = "data/pima-indians-diabetes.data.csv"
names = ["preg", "plas", "pres", "skin", "test", "mass", "pedi", "age", "class"]
dataframe = pd.read_csv(filename, names=names)
array = dataframe.values
```

```python
# separate array into input and output components
X = array[:, 0:8]
y = array[:, 8]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)

# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:5, :])
```

```python
# standardize data (mean=0, stdev=1)
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)

# summarize transformed data
np.set_printoptions(precision=3)
print(rescaledX[0:5, :])
```

```python
# normalize data (length=1)
scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)

# sum
print(normalizedX[0:5, :])
```

```python
# binarize data (set feature values to 0 or 1 depending on a threshold value)
binarizer = Binarizer(threshold=0.0).fit(X)
binarizedX = binarizer.transform(X)

# summarize transformed data
print(binarizedX[0:5, :])
```
