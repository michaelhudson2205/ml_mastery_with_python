```python
# load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
filename = 'data/pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(filename, names=names)
```

```python
# univariate histogram
data.hist(figsize=(10,8))
plt.show()
```

```python
# pip install scipy
# univariate density plot
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(10,8))
plt.show()
```

```python
# box and whisker plots
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, figsize=(10,8))
plt.show()
```

```python
# correlation matrix plot
import numpy as np
import matplotlib.colors as mcolors

# create custom colormap
# cmap = mcolors.LinearSegmentedColormap.from_list(
#     "blue_green_red",
#     [(0, "blue"), (0.5, "green"), (1, "red")])

correlations = data.corr()
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1, cmap='RdYlBu_r')
fig.colorbar(cax)
ticks = np.arange(0,len(correlations.columns),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```
